# Import libraries

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from umap import UMAP
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import shap
import re
import matplotlib; matplotlib.rcParams['figure.dpi']=300
```

# Load pathology reports

```python
path_reports=pd.read_pickle("path_report_corpus.pkl")
path_reports["X"]=pd.DataFrame.from_records(path_reports["X"]).drop(columns=["index"])
```

# Concatenate text from all reporting fields

```python
report_texts_concatenated=path_reports["X"].fillna("").apply(lambda x: " ".join(x),axis=1)
```

# Generate count matrix

```python
count_vec=CountVectorizer(stop_words="english")
count_mat=count_vec.fit_transform(report_texts_concatenated)
```

```python
count_mat_df=pd.DataFrame(count_mat.todense().A,columns=count_vec.get_feature_names())
```

```python
count_mat_df.head()
```

```python
cpt_codes=path_reports['y']
```

# Run unsupervised clustering

```python
t_data=pd.DataFrame(UMAP(n_components=2,random_state=42).fit_transform(count_mat_df),columns=['x','y'])#,metric="hellinger"
```

```python
t_data['cpt_code']=cpt_codes
t_data['high_complexity']=cpt_codes.astype(int)>88305
```

```python
sns.scatterplot(x="x",y="y",hue="cpt_code",data=t_data)
sns.despine()
```

```python
sns.scatterplot(x="x",y="y",hue="high_complexity",data=t_data)
sns.despine()
```

# Generate training and validation data

```python
X_train,X_val,y_train,y_val=train_test_split(count_mat_df,t_data['high_complexity'],stratify=t_data['high_complexity'],random_state=42)
```

# Fit a classifier to predict high complexity codes

```python
rf=RandomForestClassifier(n_estimators=5, random_state=42,class_weight="balanced").fit(X_train,y_train)
```

```python
print(classification_report(y_val,rf.predict(X_val)))
```

```python
plt.figure(figsize=(15,15))
plot_tree(rf.estimators_[0],feature_names=X_train.columns, class_names=["CPT<88305","CPT>=88305"], filled=True)
plt.show()
```

```python
plt.figure(figsize=(15,15))
plot_tree(rf.estimators_[1],feature_names=X_train.columns, class_names=["CPT<88305","CPT>=88305"], filled=True)
plt.show()
```

# Report most important words
